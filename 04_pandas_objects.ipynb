### Pandas 객체: <strong> Series </strong>

````python
# [+] List 객체에서 Series 객체를 생성
ser = pd.Series([0.25, 0.5, 0.75, 1.0])
ser
````

````python
# Pandas 객체 속성: .values
vals = ser.values
vals
# type(vals)
````

````python
# Pandas 객체 속성: .index
ind = ser.index
print(ind)
print(list(ind)) # index 값에 바로 접근 가능
````

````python
# [+] 레이블 기반 인덱싱
ser = pd.Series([0.25, 0.5, 0.75, 1.0], index = ['i', 'ii', 'iii', 'iv'])
print(ser)

print()
print(ser['iii'])
````

### <strong> Dictionary와 Series 객체 </strong>

````python
# Dictionary 객체에서 Series 객체를 생성
population_dict = {
    'California': 38332521,   # 미국 연방주 별 인구
    'Texas': 26448193,
    'New York': 19651127,
    'Florida': 19552860,
    'Illinois': 12882135
}

population = pd.Series(population_dict)
population
````

````python
# [+] 레이블 기반 인덱싱: 'California'에서 'New York'까지
population['California':'New York']
# label은 0부터 시작하지 않음 -> 'California'에서 'New York'까지 출력 됨
# 파티션 기법도 정수형 index 인지 label index 인지 구분 중요
````

### Pandas 객체:<strong> DataFrame </strong>

````python
# 미국 연방주 별 면적
area_dict = {
    'California': 423967,
    'Texas': 695662,
    'New York': 141297,
    'Florida': 170312,
    'Illinois': 149995
}

area = pd.Series(area_dict)
area
````

````python
# [+] 'population'과 'area' Series 객체를 포함하는 DataFrame 객체 생성
states = pd.DataFrame({'population' : population, 'area' : area})
states
````

````python
# DataFrame 객체의 인덱스와 컬럼
print(states.index) # 행
print(states.columns) # 열
````

````python
# [+] DataFrame으로부터 Series 객체 접근하기
states['area']
````

#### **NumPy 배열로부터 DataFrame 객체 생성**
+ 예제, 오늘의 운세: 금전운(```money_fortune```), 연애운(```love_fortune```)

````python
# [+] 실수 난수로 이루어진 12X2 크기의 NumPy 배열 생성
arr = np.random.rand(12, 2)
arr
````

````python
# NumPy 배열로부터 DataFrame 객체 생성
df = pd.DataFrame(
    arr, # 난수 배열 객체
    columns=['money_fortune', 'love_fortune'], # 열
    index=['Jan (1)', 'Feb (2)', 'Mar (3)', 'Apr (4)', 'May (5)', 'Jun (6)',
          'Jul (7)', 'Aug (8)', 'Sep (9)', 'Oct (10)', 'Nov (11)', 'Dec (12)'] # 행
)

df
````

### <strong> Series 객체 조작 </strong>

#### Dictionary 스타일 조작

````python
# Series 객체 생성
ser = pd.Series([0.25, 0.5, 0.75, 1.0], index=['a', 'b', 'c', 'd'])

print(ser)    
print('a' in ser)  # [+] Key + in 키워드 -> O
print(0.25 in ser)  # [+] Value + in 키워드 -> X // in 연산자는 key에만 적용이 가능함
print(ser.index)  # [+] 인덱스
print(ser.keys())  # [+] 키 집합
ser['e'] = 1.25  # [+] 값 추가
ser['a'] = 0.125 # [+] 값 수정
print(ser)
````

#### 배열 스타일 조작

````python
print(ser['a':'c'])   # 슬라이싱
print(ser[(ser > 0.3) & (ser < 0.8)])   # 논리 연산
# 팬시 인덱싱
ind = ['a', 'e'] 
print(ser[ind])
````

#### Pandas 객체 인덱싱
+ 정수 기반 인덱싱(암묵적, implicit)
+ 레이블 기반 인덱싱(명시적, explicit)

````python
# 슬라이싱을 이용한 'a', 'b', 'c' 선택
print(ser[0:3])   # 정수 기반 인덱싱
print(ser['a':'c'])     # 레이블 기반 인덱싱
````

### <strong> DataFrame 객체 조작 </strong>

````python
# 특정 Series 객체 접근
states['area']
# states.area
````

````python
# [+] Series 객체 추가 (밀도 = 인구 / 면적)
states['density'] = states['population'] / states['area']
states
````

#### 인덱서: ```loc```, ```iloc```

````python
# Series 객체 생성
ser = pd.Series(['a', 'b', 'c'], index=[1, 3, 5])
ser
````

````python
ser.iloc[0]
````

````python
# loc 인덱서
print(ser.loc[1])
print()
print(ser.loc[1:3])
````

````python
# iloc 인덱서
print(ser.iloc[1])
print()
print(ser.iloc[1:3])
````

````python
# DataFrame 객체 = 2차원 배열
print(states.values, '\n')        # 값들을 얻어오기
print(states.T, '\n')             # 전치행렬
print(states.iloc[:3, :2], '\n')  # 정수 기반 슬라이싱
print(states.loc[: 'Illinois', : 'population'])  # 레이블 기반 슬라이싱
````

````python
print(states.iloc[:3, :2])
````

````python
print(states.loc[: 'Illinois', : 'population'])
````

````python
# 마스킹 + 팬시 인덱싱
states.loc[states.density > 100, ['population', 'density']]
````

````python
# 값 수정
states.iloc[0, 2] = 90
states
````

````python
states.loc['Texas', 'area'] = 169000
states
````
